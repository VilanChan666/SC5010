## Train & Test Data Extraction

```python
# Importing basic libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import matplotlib.style as style
from matplotlib import colors
import warnings
sb.set()
warnings.filterwarnings("ignore")
```

```python
campaigndata = pd.read_csv("engineered_data.csv")
campaigndata.head()
```

```python
campaigndata.info()
```

```python
# OneHotEncode categorical variables via pd.get_dummies
campaigndata_encoded = pd.get_dummies(campaigndata)
campaigndata_encoded
```

```python
campaigndata_encoded.info()
```

```python
sb.catplot(y="Response", data=campaigndata_encoded, kind="count")
campaigndata_encoded['Response'].value_counts()
```

```python
# Import train_test_split from sklearn
from sklearn.model_selection import train_test_split

# Split the Dataset into Train and Test
X = campaigndata_encoded.drop(['Response'], axis=1)
y = pd.DataFrame(campaigndata_encoded['Response'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123, stratify=y)
                                                                                                                                   
# Check the sample sizes
print("Train Set :", y_train.shape, X_train.shape)
print("Test Set  :", y_test.shape, X_test.shape)

# Check class distribution 
print(y_train.value_counts(), y_test.value_counts())
sb.catplot(y="Response", data=y_train, kind="count")
sb.catplot(y="Response", data=y_test, kind="count")
```

```python
# Hybrid resampling via SMOTEENN
from imblearn.combine import SMOTEENN

sm = SMOTEENN(random_state=0)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
print(y_train_res.value_counts(), y_test.value_counts())
```

```python
train = pd.concat((X_train_res, y_train_res), axis=1)
test = pd.concat((X_test, y_test), axis=1)
```
